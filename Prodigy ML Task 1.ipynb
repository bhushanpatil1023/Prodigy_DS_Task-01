```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data
population = np.random.randint(10, 50, 20)

# Create a histogram
plt.hist(population, bins=range(10, 51, 5), align='left', rwidth=0.8)

# Set the labels
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram Example')

# Show the plot
plt.show()
```
